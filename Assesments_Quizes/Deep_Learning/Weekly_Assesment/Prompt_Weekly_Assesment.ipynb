```python
from transformers import pipeline, AutoTokenizer


print("\nLoading GPT-2 model... please wait.\n")
generator = pipeline("text-generation", model="gpt2")
print("Model loaded successfully!\n")


def generate(prompt: str, max_new_tokens: int = 150) -> str:
    full_output = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        do_sample=True,      
        temperature=0.7,        
        top_p=0.9,            
        pad_token_id=50256,     
    )[0]["generated_text"]
    return full_output[len(prompt):].strip()


article1 = """
Artificial Intelligence (AI) has become one of the most transformative technologies of the 21st century.
It is widely used in industries such as healthcare, finance, education, and manufacturing.
Machine learning, a subset of AI, allows computers to learn from data and improve performance without
explicit programming.

Recent advances in deep learning have significantly improved capabilities in image recognition,
natural language processing, and autonomous systems. For example, AI-powered diagnostic tools can
detect diseases from medical images with high accuracy.

Despite its benefits, AI also presents challenges such as data privacy concerns, ethical issues,
and job displacement. Researchers are working to create responsible AI systems that are transparent,
fair, and accountable.

In the future, AI is expected to assist humans in solving complex global problems such as climate
change, healthcare accessibility, and smart city development.
"""

article2 = """
Blockchain technology is a decentralized digital ledger used to record transactions securely.
It removes the need for intermediaries and ensures transparency and trust in digital systems.
Blockchain is widely used in cryptocurrencies, supply chain management, and digital identity
verification. Despite its promise, blockchain faces challenges such as high energy consumption,
scalability issues, and regulatory uncertainty. Researchers are exploring proof-of-stake mechanisms
and layer-2 solutions to overcome these limitations.
"""



print("  TASK 1: ARTICLE SUMMARY")


task1_prompt = (
    f"The following is a short summary of an article about artificial intelligence:\n\n"
    f"Article: {article1.strip()}\n\n"
    f"Summary:"
)

output = generate(task1_prompt, max_new_tokens=120)
print("Generated Summary:\n")
print(output)



print("  TASK 2a: ZERO-SHOT PROMPT")


zero_shot_prompt = (
    "Summarize the following article in 5 bullet points.\n\n"
    f"Article: {article1.strip()}\n\n"
    "Bullet point summary:\n•"
)

output_zero = "•" + generate(zero_shot_prompt, max_new_tokens=120)
print("Zero-Shot Output:\n")
print(output_zero)



print("  TASK 2b: FEW-SHOT PROMPT")


few_shot_prompt = (
    "Example 1:\n"
    "Article: Climate change is raising global temperatures and disrupting weather patterns.\n"
    "Summary: Global temperatures are rising due to climate change. Weather patterns are becoming "
    "more extreme. Reducing carbon emissions is essential to slow the impact.\n\n"

    "Example 2:\n"
    "Article: Online education allows students to learn remotely using digital platforms.\n"
    "Summary: Online platforms enable remote learning from anywhere. Digital tools increase "
    "accessibility for students. Technology is transforming how education is delivered.\n\n"

    "Now summarize the following article:\n"
    f"Article: {article1.strip()}\n\n"
    "Summary:"
)

output_few = generate(few_shot_prompt, max_new_tokens=130)
print("Few-Shot Output:\n")
print(output_few)




print("  TASK 2c: CHAIN-OF-THOUGHT PROMPT")


cot_prompt = (
    "Analyze the article step by step.\n\n"
    "Step 1: Identify the main topic.\n"
    "Step 2: Extract the key ideas.\n"
    "Step 3: Write a concise summary.\n\n"
    f"Article: {article1.strip()}\n\n"
    "Step 1 - Main topic: Artificial Intelligence"
)

output_cot = "Step 1 - Main topic: Artificial Intelligence" + generate(cot_prompt, max_new_tokens=140)
print("Chain-of-Thought Output:\n")
print(output_cot)


print("  TASK 3: OPTIMIZED PROMPT – Article 1")


def optimized_prompt(article: str) -> str:
    return (
        "As an academic research assistant, analyze the article below.\n\n"
        f"Article: {article.strip()}\n\n"
        "Executive Summary: This article discusses"
    )


prompt3a = optimized_prompt(article1)
output3a = "Executive Summary: This article discusses" + generate(prompt3a, max_new_tokens=150)
print("Optimized Output – Article 1:\n")
print(output3a)


print("  TASK 3: OPTIMIZED PROMPT – Article 2")


prompt3b = optimized_prompt(article2)
output3b = "Executive Summary: This article discusses" + generate(prompt3b, max_new_tokens=150)
print("Optimized Output – Article 2:\n")
print(output3b)



print("  TASK 4: TOKENIZATION EXPERIMENT")


tokenizer = AutoTokenizer.from_pretrained("gpt2")

paragraph = (
    "Artificial Intelligence enables machines to perform tasks "
    "that normally require human intelligence."
)

tokens    = tokenizer.tokenize(paragraph)
token_ids = tokenizer.encode(paragraph)

print(f"Input Paragraph:\n  {paragraph}\n")
print(f"Tokens ({len(tokens)} total):\n  {tokens}\n")
print(f"Token IDs:\n  {token_ids}\n")
print(f"Total Number of Tokens: {len(token_ids)}")



print("  TASK 5: MINI AI RESEARCH ASSISTANT")


def research_assistant(text: str) -> None:
    """
    Accepts a research article, runs three focused prompts,
    and displays a structured output: summary, insights,
    and one actionable recommendation.
    """
    print("\n-- Short Summary --")
    summary_prompt = (
        f"Article: {text.strip()}\n\n"
        "Short summary of the article:"
    )
    summary = generate(summary_prompt, max_new_tokens=100)
    print(summary)

    print("\n-- Key Insights --")
    insight_prompt = (
        f"Article: {text.strip()}\n\n"
        "Three key insights from the article:\n1."
    )
    insights = "1." + generate(insight_prompt, max_new_tokens=120)
    print(insights)

    print("\n-- Actionable Recommendation --")
    rec_prompt = (
        f"Article: {text.strip()}\n\n"
        "One actionable recommendation based on this article:"
    )
    recommendation = generate(rec_prompt, max_new_tokens=80)
    print(recommendation)



print("\nPaste a research article below.")
print("(Press Enter twice on an empty line when done)\n")

lines = []
try:
    while True:
        line = input()
        if line == "" and lines and lines[-1] == "":
            break
        lines.append(line)
    user_article = "\n".join(lines).strip()
except EOFError:
    user_article = article1.strip()

if not user_article:
    user_article = article1.strip()

print("  AI ASSISTANT OUTPUT")

research_assistant(user_article)
```
