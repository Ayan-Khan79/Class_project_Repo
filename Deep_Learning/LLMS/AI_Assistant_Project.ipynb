```python
# Scenario-Based Assessment
# Case Study: AI Assistant for a University Helpdesk

# A large university receives thousands of student queries daily related to admissions, fees, hostel facilities, course
#  schedules, and exam results.

# Currently, these queries are answered manually by staff, causing:

# Slow response time

# High operational cost

# Inconsistent responses

# The university plans to develop an AI-powered chatbot using Generative AI and Large Language Models (LLMs) to
#  automatically respond to student questions.

# You are part of the AI consulting team tasked with designing the solution.

# ==========================================
# UNIVERSITY AI HELPDESK CHATBOT
# Demonstrates:
# Generative AI
# Large Language Models (LLMs)
# Tokenization
# Transformer Inference
# Prompt Engineering
# ==========================================

# Install required libraries first
# pip install transformers torch
```

```python
from transformers import pipeline
import textwrap
from transformers import AutoTokenizer


print("Loading the chatbot. Please wait........")

chatbot = pipeline(
    "text-generation",        
    model="gpt2", 
    pad_token_id=50256,
    temperature=0.8,  
)
print("Model Loaded Successfully!!!!")

knowledge_base = """
University Helpdesk Information

Hostel Admission Requirements:
- Admission confirmation letter
- Government ID proof
- Two passport size photographs

Course Registration:
- Registration happens online through the university portal
- Students must register before the semester deadline

Fee Payment:
- Fees can be paid through the student dashboard
- Online payment methods include debit card, credit card, and net banking

Examination Schedule:
- Semester exams usually start in December and May

Library Access:
- Students must carry their university ID card
"""

def build_prompt(student_question):
    prompt = f"""
    You are a helpful university helpdesk assistant.
    Use the following information to answer the student's question accurately and concisely.

    Context:
    {knowledge_base}

    Student Question: {student_question}

    Answer:"""
    return prompt.strip()

def get_helpdesk_response(student_question):
    # Build the engineered prompt
    prompt = build_prompt(student_question)

    response = chatbot(
        prompt,
        max_new_tokens=150,   
        do_sample=False,     
    )

   # AFTER (fixed) - strips prompt, returns only the answer part
    full_output = response[0]["generated_text"].strip()
    answer = full_output.split("Answer:")[-1].strip()
    return answer if answer else "Please contact the helpdesk directly."

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")

sample_text = "What documents are needed for hostel admission?"

tokens     = tokenizer.tokenize(sample_text)
token_ids  = tokenizer.encode(sample_text)

print("=" * 55)
print("  TOKENIZATION DEMO")
print("=" * 55)
print(f"\n  Input Text : {sample_text}")
print(f"  Tokens     : {tokens}")
print(f"  Token IDs  : {token_ids}")
print(f"  Token Count: {len(tokens)}")

student_queries = [
    "What documents are needed for hostel admission?",
    "How can I pay my university fees?",
    "When do semester exams start?",
    "How do I register for courses?",
    "What do I need to access the library?",
]

print("\n" + "=" * 55)
print("  UNIVERSITY AI HELPDESK — INTERACTIVE MODE")
print("=" * 55)
print("  Ask anything about admissions, fees,")
print("  hostel, courses, exams, or library.")
print("  Type 'exit' to quit.")
print("=" * 55)

while True:
    user_input = input("\n  🎓 Your Question: ").strip()

    # Exit condition
    if user_input.lower() == "exit":
        print("\n  Thank you for using the University Helpdesk!")
        print("  Goodbye! 👋")
        break

    # Skip empty input
    if not user_input:
        print("  ⚠️  Please type a question.")
        continue

    answer = get_helpdesk_response(user_input)

    # Print only the answer — clean and short
    wrapped = textwrap.fill(answer, width=52, initial_indent="  ", subsequent_indent="  ")
    print(f"\n  💬 Answer:\n{wrapped}")
    print(f"  {'-' * 50}")
```
