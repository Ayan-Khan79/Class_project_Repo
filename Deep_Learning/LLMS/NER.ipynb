```python
# Scenario: Sports Event Analysis

# Context:

# You are part of a sports analytics team working for a global media company. Your manager has asked you to analyze match reports and press releases to identify key entities such as teams, players, locations, dates, and numbers. This information will help the company build automated summaries and highlight reels for fans.

# Task:

# You receive the following match report excerpt:

# "Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London. Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024"

# Your job is to run Named Entity Recognition (NER) using spacy to automatically extract structured insights.
```

```python

import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London.
Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024.
"""


doc = nlp(text)

print("Named Entities (text, label):")

for ent in doc.ents:
    print(f"{ent.text:25} -> {ent.label_}")


from collections import defaultdict


by_label = defaultdict(list)

for ent in doc.ents:
    by_label[ent.label_].append(ent.text)

print("\nEntities grouped by label:")

for label, items in by_label.items():
    print(f"{label}: {sorted(set(items))}")
```

```python
# 📘 Scenario: Smart Manufacturing Insights
# Context:
# You are part of a data science team in a smart manufacturing company. Your manager wants you to explore how different 
# technical concepts (like maintenance, telemetry, quality control, supply chain) are related in company documents. 
# By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are
#  connected. This helps in building recommendation systems, knowledge graphs, or even automated reporting tools.
# Task:
# Given a short corpus describing manufacturing operations, train a Word2Vec model and analyze word similarities.


# Word embeddings with gensim Word2Vec
# Setup: python -m pip install gensim nltk
```

```python
import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

nltk.download("punkt")
nltk.download("punkt_tab")

corpus = [
    "Predictive maintenance helps avoid equipment failure in the factory.",
    "Sensors collect telemetry data from machines on the production floor.",
    "Quality control ensures that defective products are removed from the supply chain.",
    "Machine learning models predict equipment downtime and schedule maintenance.",
    "IoT devices monitor temperature, pressure, and vibration in real time.",
    "Supply chain optimization reduces costs and improves delivery timelines.",
    "Defect detection using computer vision improves product quality.",
    "Telemetry data from sensors helps engineers diagnose machine failures.",
    "Automated robots perform assembly tasks on the production line.",
    "Data analytics enables smarter decisions in manufacturing operations.",
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

print("Sample Tokenized Sentences:")
for i, tokens in enumerate(tokenized_corpus[:3]):
    # Show first 3 tokenized sentences as a preview
    print(f"  Sentence {i+1}: {tokens}")


model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=100,
)

print("\nWord2Vec model trained successfully!")
print(f"Vocabulary size: {len(model.wv)} words\n")


print("Word Vector for 'maintenance' (first 5 dimensions):")
print(f"  {model.wv['maintenance'][:5]}\n")


print("Words most similar to 'maintenance':")
for word, score in model.wv.most_similar("maintenance", topn=5):
    # word  -> similar word found in vocabulary
    # score -> cosine similarity score (range: -1 to 1)
    print(f"  {word:20} -> similarity: {score:.4f}")

print("\nWords most similar to 'telemetry':")
for word, score in model.wv.most_similar("telemetry", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")

print("\nWords most similar to 'quality':")
for word, score in model.wv.most_similar("quality", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")


print("\nPairwise Word Similarity Scores:")

word_pairs = [
    ("maintenance", "failure"),
    ("sensors",     "telemetry"),
    ("quality",     "defect"),
    ("supply",      "chain"),
    ("machine",     "learning"),
]

for word1, word2 in word_pairs:
    score = model.wv.similarity(word1, word2)
    print(f"  '{word1}' <-> '{word2}': {score:.4f}")


print("\nOdd Word Out:")

groups = [
    ["maintenance", "failure", "sensors", "delivery"],
    ["telemetry",   "sensors", "IoT",     "supply"],
]

for group in groups:
    valid = [w for w in group if w in model.wv]
    if len(valid) > 1:
        odd = model.wv.doesnt_match(valid)
        print(f"  Group {valid} -> Odd one out: '{odd}'")

model.save("smart_manufacturing_word2vec.model")
print("\nModel saved as 'smart_manufacturing_word2vec.model'")

loaded_model = Word2Vec.load("smart_manufacturing_word2vec.model")
print("Model reloaded successfully!")
print("\nWords most similar to 'data' (from reloaded model):")
for word, score in loaded_model.wv.most_similar("data", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")
```

```python
# 📘 Scenario: Healthcare Research Insights
# Context:
# You are part of a data science team in a healthcare research institute. Your manager wants you to analyze medical reports and research papers to explore how different technical concepts (like diagnosis, treatment, symptoms, patients, medication) are related. By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are connected. This helps in building clinical decision support systems, medical knowledge graphs, or even automated literature reviews.
# Task:
# Given a short corpus describing healthcare operations, train a Word2Vec model and analyze word similarities.
```

```python

import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

nltk.download("punkt")
nltk.download("punkt_tab")

corpus = [
    "Diagnosis of diabetes requires blood glucose tests and patient history.",
    "Treatment for hypertension includes medication and lifestyle changes.",
    "Symptoms of pneumonia include fever, cough, and difficulty breathing.",
    "Patients with chronic kidney disease require regular dialysis sessions.",
    "Medication dosage is adjusted based on the patient's age and weight.",
    "Clinical trials evaluate the effectiveness of new drugs and therapies.",
    "MRI scans help doctors detect tumors and abnormalities in the brain.",
    "Vaccination prevents the spread of infectious diseases in communities.",
    "Surgical procedures require anesthesia, sterile instruments, and recovery care.",
    "Electronic health records store patient diagnosis, treatment, and medication history.",
    "Machine learning models predict patient readmission and disease progression.",
    "Symptoms of heart disease include chest pain, fatigue, and shortness of breath.",
    "Blood pressure monitoring is essential for hypertension and cardiac patients.",
    "Oncologists recommend chemotherapy and radiation for cancer treatment.",
    "Pediatric patients require specialized dosage and child-friendly medication.",
]


tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

print("Sample Tokenized Sentences:")
for i, tokens in enumerate(tokenized_corpus[:3]):
    # Show first 3 tokenized sentences as a preview
    print(f"  Sentence {i+1}: {tokens}")


model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=100,
)

print("\nWord2Vec model trained successfully!")
print(f"Vocabulary size: {len(model.wv)} words\n")

print("Word Vector for 'diagnosis' (first 5 dimensions):")
print(f"  {model.wv['diagnosis'][:5]}\n")


print("Words most similar to 'diagnosis':")
for word, score in model.wv.most_similar("diagnosis", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")

print("\nWords most similar to 'treatment':")
for word, score in model.wv.most_similar("treatment", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")

print("\nWords most similar to 'symptoms':")
for word, score in model.wv.most_similar("symptoms", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")

print("\nPairwise Word Similarity Scores:")

word_pairs = [
    ("diagnosis",  "treatment"),
    ("symptoms",   "disease"),
    ("medication", "dosage"),
    ("patient",    "doctor"),
    ("surgery",    "anesthesia"),
]

for word1, word2 in word_pairs:
    # Check both words exist in the trained vocabulary before comparing
    if word1 in model.wv and word2 in model.wv:
        score = model.wv.similarity(word1, word2)
        print(f"  '{word1}' <-> '{word2}': {score:.4f}")
    else:
        print(f"  '{word1}' or '{word2}' not found in vocabulary.")

print("\nOdd Word Out:")

groups = [
    ["diagnosis", "treatment", "symptoms", "vaccination"],
    ["medication", "dosage",   "patient",  "MRI"],
]

for group in groups:
    valid = [w for w in group if w in model.wv]
    if len(valid) > 1:
        odd = model.wv.doesnt_match(valid)
        print(f"  Group {valid} -> Odd one out: '{odd}'")

model.save("healthcare_word2vec.model")
print("\nModel saved as 'healthcare_word2vec.model'")

loaded_model = Word2Vec.load("healthcare_word2vec.model")
print("Model reloaded successfully!")
print("\nWords most similar to 'patient' (from reloaded model):")
for word, score in loaded_model.wv.most_similar("patient", topn=5):
    print(f"  {word:20} -> similarity: {score:.4f}")
```
