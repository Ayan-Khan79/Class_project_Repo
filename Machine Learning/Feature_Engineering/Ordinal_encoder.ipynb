```python
# Scenario: Employee Training & Satisfaction Survey
# A company conducts a survey to understand how employee education level and job satisfaction affect performance.
# They collect data such as:
# - Education: High School, Bachelor, Master, PhD
# - Satisfaction: Poor, Average, Good, Excellent
# Since these categories have a natural order (e.g., PhD is higher than Bachelor, Excellent is better
#  than Good), the company decides to use Ordinal Encoding to convert them into numbers that respect 
#  this ranking.
# They define custom ordering:
# - Education → High School (0), Bachelor (1), Master (2), PhD (3)
# - Satisfaction → Poor (0), Average (1), Good (2), Excellent (3)
# They also compare this with a manual dictionary mapping to ensure consistency.
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder


data = pd.DataFrame({
    "Education": ["High School", "Bachelor", "Master", "PhD",
                  "Bachelor", "Master"],
    "Satisfaction": ["Poor", "Good", "Excellent", "Average",
                     "Good", "Poor"]
})

print("Original Data:")
print(data)

education_order = ["High School", "Bachelor", "Master", "PhD"]
satisfaction_order = ["Poor", "Average", "Good", "Excellent"]

encoder = OrdinalEncoder(categories=[education_order, satisfaction_order])

data_encoded = data.copy()
data_encoded[["Education", "Satisfaction"]] = encoder.fit_transform(
    data[["Education", "Satisfaction"]]
)

print("\nOrdinal Encoded Data (Sklearn):")
print(data_encoded)

education_mapping = {
    "High School": 0,
    "Bachelor": 1,
    "Master": 2,
    "PhD": 3
}

satisfaction_mapping = {
    "Poor": 0,
    "Average": 1,
    "Good": 2,
    "Excellent": 3
}

data_manual = data.copy()
data_manual["Education"] = data_manual["Education"].map(education_mapping)
data_manual["Satisfaction"] = data_manual["Satisfaction"].map(satisfaction_mapping)

print("\nManual Mapping:")
print(data_manual)

print("\nAre both encodings equal?")
print(data_encoded.equals(data_manual))


```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

data = pd.DataFrame({
    "Disease_Severity": ["Mild", "Severe", "Critical", "Moderate",
                         "Severe", "Mild"],
    "Recovery_Satisfaction": ["Good", "Poor", "Excellent", "Average",
                              "Good", "Poor"]
})

print("Original Data:")
print(data)

severity_order = ["Mild", "Moderate", "Severe", "Critical"]
satisfaction_order = ["Poor", "Average", "Good", "Excellent"]

encoder = OrdinalEncoder(
    categories=[severity_order, satisfaction_order],
    dtype=int  # ensures integer output
)

data_encoded = data.copy()
data_encoded[["Disease_Severity", "Recovery_Satisfaction"]] = \
    encoder.fit_transform(
        data[["Disease_Severity", "Recovery_Satisfaction"]]
    )

print("\nOrdinal Encoded Data :")
print(data_encoded)

severity_mapping = {
    "Mild": 0,
    "Moderate": 1,
    "Severe": 2,
    "Critical": 3
}

satisfaction_mapping = {
    "Poor": 0,
    "Average": 1,
    "Good": 2,
    "Excellent": 3
}

data_manual = data.copy()
data_manual["Disease_Severity"] = \
    data_manual["Disease_Severity"].map(severity_mapping)

data_manual["Recovery_Satisfaction"] = \
    data_manual["Recovery_Satisfaction"].map(satisfaction_mapping)

print("\nManual Mapping:")
print(data_manual)

print("\nAre both encodings equal?")
print(data_encoded.equals(data_manual))

```
