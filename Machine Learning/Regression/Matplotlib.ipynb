```python
# ðŸ“˜ Scenario Question # A college is tracking student attendance across the first four months of the semester: # - January: 85 students present on average # - February: 90 students # - March: 95 students # - April: 88 students # The academic coordinator wants to visualize the attendance trend to see whether students are attending classes more regularly or # if there are fluctuations. # You are asked to: # - Plot a line chart showing the monthly attendance trend. # - Label the axes and add a title to make the chart clear. # - Interpret the chart: # - Which month had the highest attendance? # - Did attendance steadily increase, or were there ups and downs? # - What might explain the dip in April?
```

```python
import matplotlib.pyplot as plt
months = ["January", "February", "March", "April"]
attendance = [85, 90, 95, 88]

plt.plot(months, attendance, marker='o')

# Labels and Title
plt.xlabel("Months")
plt.ylabel("Average Students Present")
plt.title("Monthly Student Attendance Trend")

plt.grid(True)
plt.show()
```
