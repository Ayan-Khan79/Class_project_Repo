### Scenario: AI Marketing Content & Validation System (Corporate Marketing Team)
🎯 Problem Statement
In companies:
- Marketing teams draft campaign content manually.
- Editors check for tone, compliance, and brand alignment.
- Testing across platforms (emails, social posts, ads) is manual and time-consuming.
👉 Solution
A multi-agent system where:
- ✍️ Content Creator Agent generates marketing copy, visuals, or ad text.
- 🔎 Validator Agent reviews for compliance, tone, and brand guidelines, then simulates deployment.
- 🔁 The loop continues until the content meets quality and compliance standards.
🔄 Workflow
Campaign Request → AI drafts content → Validator checks tone/compliance → Test deployment simulation → Iteration loop → Final approved campaign assets
🧹 Conflict Resolution
- Resolve conflicting brand guidelines (e.g., tone mismatch between product lines).
- Standardize formatting across channels (social, email, ads).
- Ensure package/library consistency for creative assets (e.g., Canva API vs. Adobe API).

```python
# Scenario: AI Marketing Content & Validation System
# Characters:
# - Content_Creator (Assistant Agent) → Drafts marketing copy, ad text, and campaign content.
# - Validator (Assistant Agent) → Reviews for tone, compliance, and brand guidelines.
# - Campaign_Manager (UserProxy Agent) → Orchestrates the workflow and runs the loop.

import os
import autogen
from dotenv import load_dotenv

# 🔑 Step 1: Set API Key
load_dotenv()
GROQ_API_KEY = os.getenv('GROQ_API_KEY')

# 🔧 Step 2: Configure Groq LLM
config_list = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key": GROQ_API_KEY,
        "base_url": "https://api.groq.com/openai/v1"
    }
]
llm_config = {"config_list": config_list, "temperature": 0.7}

# ✍️ Step 3: Create Content Creator Agent
content_creator = autogen.AssistantAgent(
    name="Content_Creator",
    llm_config=llm_config,
    system_message="""
    You are an expert marketing content creator for a corporate brand.
    Your job is to:
    - Draft compelling marketing copy for emails, social posts, and ads.
    - Follow brand guidelines: professional tone, inclusive language, clear CTA.
    - Format content separately for each platform:
        * EMAIL: Subject line + body
        * SOCIAL: Short punchy post under 280 chars
        * AD: Headline + tagline + CTA
    - Revise content based on validator feedback.
    Always label each platform section clearly.
    """
)

# 🔎 Step 4: Create Validator Agent
validator = autogen.AssistantAgent(
    name="Validator",
    llm_config=llm_config,
    system_message="""
    You are a strict marketing compliance and brand validator.
    Your job is to review content for:
    - ✅ Tone: Must be professional, positive, and inclusive.
    - ✅ Compliance: No false claims, no offensive language, no legal risks.
    - ✅ Brand Alignment: Consistent voice across email, social, and ads.
    - ✅ Formatting: Each platform must follow its format standards.
    - ✅ CTA: Every piece must have a clear Call-To-Action.

    Conflict Resolution Rules:
    - If tone mismatches between platforms → flag and request revision.
    - If formatting is inconsistent → specify exact corrections needed.
    - If brand guidelines conflict → default to professional and inclusive tone.

    Respond with:
    - APPROVED ✅ if content meets all standards.
    - REVISION NEEDED ❌ with specific feedback if it does not.
    - Always give a compliance score out of 10.
    """
)

# ⚙️ Step 5: Create Campaign Manager (UserProxy)
campaign_manager = autogen.UserProxyAgent(
    name="Campaign_Manager",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,  # Allow up to 5 revision loops
    code_execution_config={"work_dir": "campaign_output", "use_docker": False},
    is_termination_msg=lambda msg: "APPROVED ✅" in msg.get("content", "")
)

# 🔁 Step 6: Set up Group Chat for Multi-Agent Loop
group_chat = autogen.GroupChat(
    agents=[campaign_manager, content_creator, validator],
    messages=[],
    max_round=15,  # Maximum rounds of iteration
    speaker_selection_method="round_robin"  # Creator → Validator → Manager loop
)

group_chat_manager = autogen.GroupChatManager(
    groupchat=group_chat,
    llm_config=llm_config
)

# 🚀 Step 7: Launch Campaign Workflow
campaign_manager.initiate_chat(
    group_chat_manager,
    message="""
    📢 New Campaign Request:

    Product: AI-Powered Productivity Suite
    Target Audience: Corporate professionals aged 25-45
    Campaign Goal: Drive free trial signups
    Brand Guidelines:
        - Tone: Professional, innovative, empowering
        - Avoid: Overpromising, technical jargon, negative comparisons
        - CTA: "Start Your Free Trial Today"

    Please generate marketing content for:
    1. 📧 EMAIL — Subject line + body (max 150 words)
    2. 📱 SOCIAL POST — Twitter/LinkedIn (max 280 chars)
    3. 📣 AD COPY — Headline + tagline + CTA

    Validator will review each piece for:
    - Tone & compliance
    - Brand alignment
    - Platform formatting
    - Conflict resolution between platforms

    Iterate until all content is APPROVED ✅
    """
)
```

### # Scenario: AI-Powered Dev Workflow (Corporate Engineering Team)
🎯 Problem Statement In companies: - Developers manually write and test code. - Reviewers check for bugs and compliance. - Testing and validation are slow and error-prone. 👉 Solution A multi-agent system powered by Groq API where: - 👨‍💻 Coder Agent writes Python code for requested features. - 🔍 Reviewer Agent validates logic and ensures best practices. - 🧪 Tester Agent generates unit tests and executes them automatically. - 🔁 The loop continues until the code passes all tests and meets requirements. 🔄 Workflow Feature Request → Product Manager breaks into tasks → Coder writes code → Tester validates with unit tests → Reviewer checks compliance → Iteration loop → Final approved codebase 
🧹 Conflict Resolution  
 - Resolve conflicting dependencies (e.g., package version mismatches).
 - Standardize coding style across agents (PEP8 compliance).
 - Ensure Groq API integration avoids latency or model mismatch issues.

```python
# Scenario: AI-Powered Dev Workflow (Corporate Engineering Team)
# Characters:
# - Product_Manager (UserProxy Agent) → Breaks feature requests into tasks and orchestrates workflow.
# - Coder (Assistant Agent) → Writes Python code for requested features.
# - Reviewer (Assistant Agent) → Validates logic, best practices, and PEP8 compliance.
# - Tester (Assistant Agent) → Generates and executes unit tests automatically.

import os
import autogen
from dotenv import load_dotenv

# 🔑 Step 1: Set API Key
load_dotenv()
GROQ_API_KEY = os.getenv('GROQ_API_KEY')

# 🔧 Step 2: Configure Groq LLM
config_list = [
    {
        "model": "llama-3.1-8b-instant",
        "api_key": GROQ_API_KEY,
        "base_url": "https://api.groq.com/openai/v1"
    }
]
llm_config = {"config_list": config_list, "temperature": 0}

# 👨‍💻 Step 3: Create Coder Agent
coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message="""
    You are an expert Python developer in a corporate engineering team.
    Your job is to:
    - Write clean, efficient, and well-documented Python code.
    - Follow PEP8 coding standards strictly.
    - Add docstrings to every function and class.
    - Handle edge cases and exceptions properly.
    - Resolve dependency conflicts by using widely compatible package versions.
    - Revise code based on Reviewer and Tester feedback.

    Conflict Resolution Rules:
    - If package version conflicts arise → use the most stable LTS version.
    - If logic conflicts with requirements → ask Product_Manager for clarification.
    - Always wrap code in proper Python code blocks.
    """
)

# 🔍 Step 4: Create Reviewer Agent
reviewer = autogen.AssistantAgent(
    name="Reviewer",
    llm_config=llm_config,
    system_message="""
    You are a senior code reviewer in a corporate engineering team.
    Your job is to review code for:
    - ✅ Logic correctness and edge case handling.
    - ✅ PEP8 compliance (naming, spacing, line length).
    - ✅ Security vulnerabilities (SQL injection, hardcoded secrets).
    - ✅ Performance issues (unnecessary loops, memory leaks).
    - ✅ Documentation (docstrings, comments, type hints).
    - ✅ Dependency consistency (no version mismatches).

    Conflict Resolution Rules:
    - If coding style conflicts with PEP8 → always enforce PEP8.
    - If security risk found → flag as CRITICAL and block approval.
    - If minor issues found → suggest improvements but allow progression.

    Respond with:
    - CODE APPROVED ✅ if code meets all standards.
    - REVISION NEEDED ❌ with specific line-by-line feedback if it does not.
    - Always give a code quality score out of 10.
    """
)

# 🧪 Step 5: Create Tester Agent
tester = autogen.AssistantAgent(
    name="Tester",
    llm_config=llm_config,
    system_message="""
    You are an expert QA engineer in a corporate engineering team.
    Your job is to:
    - Generate comprehensive unit tests using pytest.
    - Cover all edge cases: empty inputs, negative numbers, type errors, boundary values.
    - Write tests BEFORE checking if they pass (TDD approach).
    - Execute tests and report results clearly.
    - Ensure minimum 80% code coverage.

    Test Structure:
    - Group tests by function name.
    - Use descriptive test names: test_function_name_scenario().
    - Include both positive and negative test cases.
    - Use pytest fixtures where appropriate.

    Conflict Resolution Rules:
    - If tests conflict with requirements → flag to Product_Manager.
    - If test environment dependencies mismatch → specify exact versions needed.

    Respond with:
    - ALL TESTS PASSED ✅ with coverage report if all tests pass.
    - TESTS FAILED ❌ with specific failure details if tests fail.
    - Always report total tests run, passed, and failed.
    """
)

# ⚙️ Step 6: Create Product Manager (UserProxy)
product_manager = autogen.UserProxyAgent(
    name="Product_Manager",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=15,
    code_execution_config={
        "work_dir": "dev_output",
        "use_docker": False
    },
    is_termination_msg=lambda msg: (
        "CODE APPROVED ✅" in msg.get("content", "") and
        "ALL TESTS PASSED ✅" in msg.get("content", "")
    )
)

# 🔁 Step 7: Set up Group Chat
group_chat = autogen.GroupChat(
    agents=[product_manager, coder, tester, reviewer],
    messages=[],
    max_round=20,
    speaker_selection_method="round_robin"  # Manager → Coder → Tester → Reviewer loop
)

group_chat_manager = autogen.GroupChatManager(
    groupchat=group_chat,
    llm_config=llm_config
)

# 🚀 Step 8: Launch Dev Workflow
product_manager.initiate_chat(
    group_chat_manager,
    message="""
    🚀 New Feature Request:

    Feature: Student Grade Management System
    Language: Python
    Requirements:
        1. Class Student with attributes: name, grades (list of integers)
        2. Method calculate_average() → returns average grade
        3. Method get_highest_grade() → returns highest grade
        4. Method get_lowest_grade() → returns lowest grade
        5. Method get_grade_status() → returns Pass if average >= 50 else Fail
        6. Handle edge cases: empty grades list, invalid inputs

    Coding Standards:
        - PEP8 compliance
        - Type hints on all functions
        - Docstrings on all methods
        - No hardcoded values

    Testing Requirements:
        - Unit tests with pytest
        - Test all methods
        - Test edge cases
        - Minimum 80% coverage

    Workflow:
        1. Coder writes the Student class
        2. Tester generates and runs unit tests
        3. Reviewer checks code quality and PEP8
        4. Loop until CODE APPROVED ✅ and ALL TESTS PASSED ✅
    """
)
```
