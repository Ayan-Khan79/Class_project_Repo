````python
# Scenario: AI Content Automation for a Startup
# 🏢 Problem Statement

# A startup wants to:

# Stay updated with latest AI trends

# Automatically create LinkedIn posts

# Save time on manual research & writing

# 👉 So we build a multi-agent AI system to automate this using CrewAI.

import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from dotenv import load_dotenv
from langchain_litellm import ChatLiteLLM # Updated import

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

# ❌ REMOVE any OPENAI key completely
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
````

````python
# Scenario: Automated Learning Hub for an EdTech Company
# 🎓 Problem Statement
# An EdTech company wants to:
# - Keep track of the latest educational technology innovations
# - Automatically generate weekly blog posts for their website
# - Reduce time spent on manual research and content drafting
# 👉 So we build a multi-agent AI system using CrewAI to automate this workflow:
# - Trend Monitoring Agent → Continuously scans for new EdTech tools, teaching strategies, and AI-in-education updates
# - Content Summarization Agent → Condenses research into digestible insights tailored for educators and learners
# - Blog Writing Agent → Crafts engaging, SEO-friendly blog posts automatically
# - Publishing Agent → Schedules and posts content directly to the company’s website
````

````python
import os
import json
from datetime import datetime
from dotenv import load_dotenv

# ── Load environment ──────────────────────────────────────────
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")


os.environ["GROQ_API_KEY"] = GROQ_API_KEY
os.environ["CREWAI_TRACING_ENABLED"] = "false"   # set "true" to enable LangSmith tracing

# ── Imports ───────────────────────────────────────────────────
from crewai import Agent, Task, Crew, Process
# ── LLM Setup (Groq — LLaMA 3.3 70B) ─────────────────────────
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")



# ════════════════════════════════════════════════════════════════
# AGENTS
# ════════════════════════════════════════════════════════════════

trend_monitor = Agent(
    role="EdTech Trend Monitor",
    goal=(
        "Continuously scan and identify the latest innovations in educational technology, "
        "including AI-powered learning tools, adaptive platforms, immersive tech (AR/VR), "
        "gamification, new pedagogical strategies, and government/policy shifts in EdTech."
    ),
    backstory=(
        "You are a seasoned EdTech analyst who reads research papers, follows top educators "
        "on social media, monitors product launches from companies like Duolingo, Coursera, "
        "Khan Academy, Google for Education, and tracks emerging startups. "
        "You have a sharp eye for what's trending vs. what's just hype."
    ),
    llm=llm,
    verbose=True,
    allow_delegation=False,
    max_iter=4,
)

content_summarizer = Agent(
    role="Educational Content Summarizer",
    goal=(
        "Transform raw research findings and trend reports into clear, accurate, "
        "and engaging summaries tailored for educators, instructional designers, and learners."
    ),
    backstory=(
        "You hold a Master's in Instructional Design and have spent 8 years writing for "
        "EdTech publications like EdSurge and eLearning Industry. You excel at breaking down "
        "complex research into practical takeaways without losing accuracy. "
        "You always cite sources and highlight 'why it matters for teachers'."
    ),
    tools=[],
    llm=llm,
    verbose=True,
    allow_delegation=False,
    max_iter=3,
)

blog_writer = Agent(
    role="SEO-Optimised Blog Writer",
    goal=(
        "Write compelling, SEO-friendly blog posts that rank on Google, engage educators, "
        "and drive traffic to the EdTech company's website."
    ),
    backstory=(
        "You are a content strategist and writer who has grown multiple EdTech blogs to "
        "100k+ monthly readers. You understand on-page SEO, know how to craft irresistible "
        "headlines, use structured headers (H1/H2/H3), include relevant keywords naturally, "
        "and always end with a strong call-to-action. You write in a warm, conversational tone."
    ),
    tools=[],
    llm=llm,
    verbose=True,
    allow_delegation=False,
    max_iter=3,
)

publisher = Agent(
    role="Content Publishing Coordinator",
    goal=(
        "Prepare the final blog post for publication: format it cleanly in Markdown, "
        "add SEO metadata (title, description, keywords, slug), suggest a publishing schedule, "
        "and generate a ready-to-deploy JSON payload for the CMS API."
    ),
    backstory=(
        "You are a digital publishing expert who has managed content pipelines for major "
        "online publications. You know WordPress REST API, Contentful, Strapi, and Ghost CMS. "
        "You ensure every post has proper metadata, is mobile-friendly in structure, "
        "and hits the optimal posting window for the target audience (educators, 7–9 AM weekdays)."
    ),
    tools=[],
    llm=llm,
    verbose=True,
    allow_delegation=False,
    max_iter=3,
)


# ════════════════════════════════════════════════════════════════
# TASKS
# ════════════════════════════════════════════════════════════════

week_label = datetime.now().strftime("Week of %B %d, %Y")

task_monitor = Task(
    description=(
        f"Research and compile the TOP 5 EdTech trends for {week_label}. "
        "For each trend include:\n"
        "  • Trend name and one-line description\n"
        "  • Key product/tool/study driving it\n"
        "  • Why it matters for educators and learners\n"
        "  • A credible source or example (company, paper, or announcement)\n\n"
        "Focus areas: AI in education, personalised learning, immersive tech (AR/VR), "
        "gamification, microlearning, EdTech policy updates."
    ),
    expected_output=(
        "A structured report with exactly 5 numbered trends. "
        "Each trend: name, description, key driver, educator impact, source. "
        "Total length: 400–600 words."
    ),
    agent=trend_monitor,
)

task_summarize = Task(
    description=(
        "Using the trend report from the Trend Monitor, create a concise summary package:\n"
        "  1. Executive Summary (2–3 sentences covering the overall theme of this week's trends)\n"
        "  2. Educator Takeaways (bullet list — what teachers & trainers should do with each trend)\n"
        "  3. Learner Impact (how these trends improve the student/learner experience)\n"
        "  4. Jargon Glossary (define any technical terms used in the report)\n\n"
        "Tone: professional yet accessible. Audience: K-12 teachers, corporate trainers, "
        "instructional designers."
    ),
    expected_output=(
        "A 4-section summary document: Executive Summary, Educator Takeaways (5 bullets), "
        "Learner Impact (3–4 sentences), Glossary (3–5 terms defined). ~300 words."
    ),
    agent=content_summarizer,
    context=[task_monitor],
)

task_write_blog = Task(
    description=(
        "Write a full-length, SEO-optimised blog post using the trend report and summary. "
        "Structure:\n"
        "  • H1: Attention-grabbing title with primary keyword\n"
        "  • Introduction: Hook + why this matters (100 words)\n"
        "  • H2 sections: One per trend (150–200 words each) with subheadings\n"
        "  • 'What This Means For You' section: Practical advice for educators (150 words)\n"
        "  • Conclusion + Call-to-Action: Subscribe to newsletter / explore platform\n\n"
        "SEO requirements:\n"
        "  - Primary keyword: 'EdTech trends [current year]'\n"
        "  - Secondary keywords: AI in education, personalised learning, classroom technology\n"
        "  - Natural keyword density (1–2%)\n"
        "  - Include internal link placeholders [LINK: relevant-article]\n"
        "  - Estimated read time at the top\n"
        "  - Tone: conversational, inspiring, authoritative"
    ),
    expected_output=(
        "A complete blog post in Markdown format. "
        "Length: 900–1200 words. Proper H1/H2/H3 hierarchy. "
        "Includes estimated read time, introduction, 5 trend sections, "
        "practical takeaways, and CTA conclusion."
    ),
    agent=blog_writer,
    context=[task_monitor, task_summarize],
)

task_publish = Task(
    description=(
        "Take the final blog post and prepare it for CMS publishing:\n"
        "  1. Clean and validate the Markdown formatting\n"
        "  2. Generate SEO metadata:\n"
        "       - meta_title (55–60 chars)\n"
        "       - meta_description (150–160 chars)\n"
        "       - slug (URL-friendly, hyphenated)\n"
        "       - focus_keyword\n"
        "       - tags (5–7 relevant tags)\n"
        "       - category\n"
        "  3. Suggest an image alt-text for the featured image\n"
        "  4. Recommend publish datetime (next weekday, 8:00 AM IST)\n"
        "  5. Output a JSON publishing payload that can be sent to a REST CMS API\n\n"
        "The JSON should follow this schema:\n"
        "  { title, slug, meta_description, focus_keyword, tags, category,\n"
        "    status, scheduled_at, author, featured_image_alt, content }"
    ),
    expected_output=(
        "Two deliverables:\n"
        "  A) The final cleaned blog post in Markdown\n"
        "  B) A valid JSON publishing payload (wrapped in ```json ... ```) "
        "ready for CMS API submission."
    ),
    agent=publisher,
    context=[task_write_blog],
)


# ════════════════════════════════════════════════════════════════
# CREW
# ════════════════════════════════════════════════════════════════

edtech_crew = Crew(
    agents=[trend_monitor, content_summarizer, blog_writer, publisher],
    tasks=[task_monitor, task_summarize, task_write_blog, task_publish],
    process=Process.sequential,   # tasks run one after another (waterfall pipeline)
    verbose=True,
)


# ════════════════════════════════════════════════════════════════
# RUN
# ════════════════════════════════════════════════════════════════

def save_outputs(result: str):
    """Save blog post (Markdown) and CMS payload (JSON) to disk."""
    os.makedirs("outputs", exist_ok=True)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M")

    # ── Save full result ──
    full_path = f"outputs/full_result_{timestamp}.txt"
    with open(full_path, "w", encoding="utf-8") as f:
        f.write(result)

    # ── Extract and save Markdown blog ──
    md_start = result.find("# ")
    if md_start != -1:
        # Grab everything from the first H1 up to the JSON block
        json_start = result.find("```json")
        md_content = result[md_start:json_start].strip() if json_start != -1 else result[md_start:].strip()
        md_path = f"outputs/blog_post_{timestamp}.md"
        with open(md_path, "w", encoding="utf-8") as f:
            f.write(md_content)
        print(f"\nBlog post saved → {md_path}")

    # ── Extract and save JSON payload ──
    json_start = result.find("```json")
    json_end   = result.find("```", json_start + 7)
    if json_start != -1 and json_end != -1:
        raw_json = result[json_start + 7: json_end].strip()
        try:
            payload = json.loads(raw_json)
            json_path = f"outputs/cms_payload_{timestamp}.json"
            with open(json_path, "w", encoding="utf-8") as f:
                json.dump(payload, f, indent=2, ensure_ascii=False)
            print(f" CMS payload saved  → {json_path}")
        except json.JSONDecodeError:
            # Save raw even if parsing fails
            raw_path = f"outputs/cms_payload_raw_{timestamp}.txt"
            with open(raw_path, "w", encoding="utf-8") as f:
                f.write(raw_json)
            print(f"  JSON parse error — raw payload saved → {raw_path}")

    print(f" Full result saved   → {full_path}")


if __name__ == "__main__":
    print("\n" + "═" * 60)
    print("    AUTOMATED LEARNING HUB — Starting Crew")
    print("═" * 60 + "\n")

    result = edtech_crew.kickoff()

    print("\n" + "═" * 60)
    print("    CREW FINISHED — Saving outputs")
    print("═" * 60 + "\n")

    final_text = str(result)
    save_outputs(final_text)

    print("\n── FINAL OUTPUT PREVIEW ──────────────────────────────────\n")
    print(final_text[:2000] + "\n...[truncated — see outputs/ folder for full content]")
````

````python
# Scenario: Smart Job Application Assistant
#  Problem Statement

# Students struggle to:

# Analyze job descriptions

# Match their skills

# Write tailored resumes & cover letters

# 👉 So we build a multi-agent AI system that automates job application preparation.

# 🤖 System Overview
# 🔹 Agent 1: Job Analyzer

# Reads job description

# Extracts required skills

# 🔹 Agent 2: Resume Optimizer

# Matches candidate skills with job

# Suggests improvements

# 🔹 Agent 3: Cover Letter Writer

# Generates a personalized cover letter

# 🔄 Workflow
# Job Description → Skill Analysis → Resume Optimization → Cover Letter Generation


import os
from crewai import Agent, Task, Crew

from dotenv import load_dotenv

# -----------------------------
# 1. API Key Setup
# -----------------------------
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

# -----------------------------
# 2. Define Agents
# -----------------------------
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Input Data (Scenario)
# -----------------------------
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True   # good for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
````

````python
# Scenario: Automated Business Intelligence (BI) Report Generator
# 🎯 Problem Statement

# In companies:

# Managers struggle to analyze large sales data

# Reports take hours to prepare

# Insights are often delayed

# 👉 Solution: Build a multi-agent AI system that:

# Analyzes business data

# Extracts insights

# Generates executive reports

# 🤖 System Overview
# 🔹 Agent 1: Data Analyst

# Analyzes sales/business data

# Finds trends, patterns, KPIs

# 🔹 Agent 2: Business Strategist

# Interprets insights

# Suggests business decisions

# 🔹 Agent 3: Report Generator

# Creates a professional executive report


!pip install crewai litellm

import os
from crewai import Agent, Task, Crew
from dotenv import load_dotenv

# -----------------------------
# 1. API Key Setup
# -----------------------------
load_dotenv()
GROQ_API_KEY = os.getenv('GROQ_API_KEY')
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

# -----------------------------
# 2. Define Agents
# -----------------------------
data_analyst = Agent(
    role="Data Analyst",
    goal="Analyze business data and extract meaningful insights",
    backstory=(
        "You are a senior data analyst working in a corporate environment. "
        "You specialize in identifying trends, KPIs, and patterns in sales data."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

business_strategist = Agent(
    role="Business Strategist",
    goal="Provide strategic recommendations based on data insights",
    backstory=(
        "You are a business consultant who translates data insights into "
        "actionable strategies for company growth."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

report_generator = Agent(
    role="Executive Report Writer",
    goal="Generate a professional business report",
    backstory=(
        "You are an expert in corporate communication who prepares executive-level "
        "reports with clarity and professionalism."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Sample Business Data
# -----------------------------
sales_data = """
Region: North | Revenue: 12000 | Cost: 8000
Region: South | Revenue: 15000 | Cost: 9000
Region: East  | Revenue: 10000 | Cost: 7000
Region: West  | Revenue: 18000 | Cost: 11000
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following sales data and provide:
    - Total revenue and profit
    - Best performing region
    - Key trends and observations

    Data:
    {sales_data}
    """,
    expected_output="Detailed analysis with KPIs and trends",
    agent=data_analyst
)

strategy_task = Task(
    description="""
    Based on the analysis:
    - Suggest 3 business strategies
    - Identify growth opportunities
    - Highlight risks
    """,
    expected_output="Strategic recommendations",
    agent=business_strategist,
    context=[analysis_task]
)

report_task = Task(
    description="""
    Create a professional executive report including:
    - Summary
    - Key insights
    - Recommendations
    - Conclusion
    """,
    expected_output="Well-structured business report",
    agent=report_generator,
    context=[analysis_task, strategy_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[data_analyst, business_strategist, report_generator],
    tasks=[analysis_task, strategy_task, report_task],
    verbose=True,
    tracing=True   # great for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n📊 FINAL BUSINESS REPORT:\n")
print(result)
````
