# MCP SCENARIO: “Smart HR Onboarding Assistant”
## 🧩 Scenario Background
You are working in a company called XYZ Corp.
New employees often face challenges during onboarding, such as:
- Trouble accessing payroll portal
- Confusion about leave policies
- Difficulty setting up email accounts
- Questions about training schedules
👉 Instead of emailing HR or waiting for responses, employees use an AI Onboarding Bot.

🤖 What this Bot Should Do
When a new hire types a question/problem:
- Understand the query (e.g., “I can’t log into payroll”)
- Decide if escalation to HR is needed
- Identify:
- Category (Payroll / Policy / IT Setup / Training)
- Priority (High / Medium)
- Create a support ticket if required
- Provide instant guidance (FAQs, step-by-step instructions)
- Show confirmation and next steps

This way, the MCP framework is reused in a Human Resources context, where the AI assistant streamlines onboarding, reduces HR workload, and ensures employees feel supported from day one.

### Without LLM

```python
# ============================================
# STEP 0 & 1: DATABASE & TOOL
# ============================================
hr_tickets = []

def create_hr_ticket(issue, category, priority):
    t_id = f"HR-{2000 + len(hr_tickets)}"
    ticket = {
        "id": t_id, 
        "issue": issue, 
        "category": category, 
        "priority": priority
    }
    hr_tickets.append(ticket)
    return ticket

# ============================================
# STEP 2: LOGIC (Hard-coded Rules)
# ============================================
def analyze_hr_issue_rules(user_input):
    ui = user_input.lower()
    
    # Simple Keyword Matching
    if "payroll" in ui or "salary" in ui:
        return {"create": True, "cat": "Payroll", "prio": "High"}
    elif "leave" in ui or "policy" in ui:
        return {"create": True, "cat": "Policy", "prio": "Low"}
    elif "email" in ui or "login" in ui:
        return {"create": True, "cat": "IT Setup", "prio": "Medium"}
    elif "training" in ui or "schedule" in ui:
        return {"create": True, "cat": "Training", "prio": "Medium"}
    
    return {"create": False}

# ============================================
# STEP 3 & 4: AGENT AND RUN LOOP
# ============================================
print("🏢 HR Assistant (Rule-Based) Started - Type 'exit' to quit\n")

while True:
    user_input = input("New Hire Issue: ").strip()
    
    if user_input.lower() == "exit":
        print("👋 Exiting HR System...")
        break
    
    if not user_input:
        continue
    
    # 1. Decision Logic
    decision = analyze_hr_issue_rules(user_input)
    
    if decision["create"]:
        # 2. Trigger the Tool
        res = create_hr_ticket(user_input, decision["cat"], decision["prio"])
        
        # 3. Print the Full Ticket Details
        print("\n-----------------------------------------")
        print(f"✅ TICKET GENERATED SUCCESSFULLY")
        print("-----------------------------------------")
        print(f"ID       : {res['id']}")
        print(f"Category : {res['category']}")
        print(f"Priority : {res['priority']}")
        print(f"Issue    : {res['issue']}")
        print("-----------------------------------------\n")
        
    else:
        print("🤖 AI: I couldn't categorize this. Please check the FAQ or contact hr@xyzcorp.com.\n")
```

### With LLM

```python
import os, json
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv('GROQ_API_KEY'), base_url="https://api.groq.com/openai/v1")

# STEP 0 & 1: Database and Tool
hr_tickets = []

def create_hr_ticket(issue, category, priority):
    t_id = f"HR-{2000 + len(hr_tickets)}"
    ticket = {"id": t_id, "issue": issue, "category": category, "priority": priority}
    hr_tickets.append(ticket)
    return ticket

# STEP 2: LLM Analysis
def analyze_with_llm(user_input):
    prompt = f"""Categorize this HR issue: "{user_input}"
    Return JSON: {{"create_ticket": true/false, "category": "Payroll/Policy/IT/Training", "priority": "High/Medium/Low"}}"""
    
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        response_format={"type": "json_object"}
    )
    return json.loads(response.choices[0].message.content)

# ============================================
# STEP 3: MCP AGENT (Updated to Print Full Ticket)
# ============================================

def hr_mcp_agent(user_input):
    # 1. Ask the LLM to decide
    decision = analyze_with_llm(user_input)
    
    print(f"🤖 AI Decision: {decision}")

    if decision["create_ticket"]:
        # 2. Call the Tool
        res = create_hr_ticket(
            issue=user_input, 
            category=decision["category"], 
            priority=decision["priority"]
        )
        
        # 3. Print and Return the full ticket details
        return f"""
-----------------------------------------
✅ NEW HR TICKET CREATED
-----------------------------------------
Ticket ID  : {res['id']}
Category   : {res['category']}
Priority   : {res['priority']}
Description: {res['issue']}
-----------------------------------------
Status: Sent to HR Queue.
"""
    else:
        return "🤖 AI Response: No ticket needed. Please refer to the Employee Handbook for this issue."

# ============================================
# STEP 4: RUN LOOP
# ============================================

print("🚀 Smart HR Onboarding Assistant Started (type 'exit')\n")

while True:
    user_input = input("How can we help you today? ").strip()

    if user_input.lower() == "exit":
        print("👋 Goodbye! Have a great day at XYZ Corp.")
        break

    if not user_input:
        continue

    # Run the agent and print the returned string
    result_message = hr_mcp_agent(user_input)
    print(result_message)
```

### Without LLM

```python
# STEP 0 & 1: Database and Tool
banking_tickets = []

def create_banking_ticket(issue, category, priority):
    t_id = f"BANK-{5000 + len(banking_tickets)}"
    ticket = {"id": t_id, "issue": issue, "category": category, "priority": priority}
    banking_tickets.append(ticket)
    return ticket

# STEP 2: Logic (Hard-coded Rules)
def analyze_banking_rules(user_input):
    ui = user_input.lower()
    if "card" in ui or "declined" in ui:
        return {"create": True, "cat": "Card Services", "prio": "High"}
    elif "login" in ui or "password" in ui:
        return {"create": True, "cat": "Online Banking", "prio": "Medium"}
    elif "loan" in ui or "mortgage" in ui:
        return {"create": True, "cat": "Loans", "prio": "Low"}
    elif "charge" in ui or "dispute" in ui:
        return {"create": True, "cat": "Transactions", "prio": "High"}
    return {"create": False}

# STEP 3 & 4: Agent and Run Loop
print("🏦 FinTrust Bank Support (Rule-Based) - Type 'exit' to quit\n")
while True:
    user_input = input("How can we help with your account? ").strip()
    if user_input.lower() == "exit": break
    if not user_input: continue
    
    decision = analyze_banking_rules(user_input)
    if decision["create"]:
        res = create_banking_ticket(user_input, decision["cat"], decision["prio"])
        print("\n--- 🎫 BANKING TICKET GENERATED ---")
        print(f"ID      : {res['id']}\nCategory: {res['category']}\nPriority: {res['priority']}\nIssue   : {res['issue']}")
        print("----------------------------------\n")
    else:
        print("🤖 AI: Please visit our branch or check our FAQ online.\n")
```

### With LLM

```python
import os, json
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv('GROQ_API_KEY'), base_url="https://api.groq.com/openai/v1")

banking_tickets = []

def create_banking_ticket(issue, category, priority):
    t_id = f"BANK-{5000 + len(banking_tickets)}"
    ticket = {"id": t_id, "issue": issue, "category": category, "priority": priority}
    banking_tickets.append(ticket)
    return ticket

def analyze_banking_llm(user_input):
    prompt = f"""You are a Banking Assistant. Categorize this issue: "{user_input}"
    Return JSON ONLY: {{"create": true, "category": "Card Services/Online Banking/Loans/Transactions", "priority": "High/Medium/Low"}}"""
    
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        response_format={"type": "json_object"}
    )
    return json.loads(response.choices[0].message.content)

print("🚀 FinTrust AI Support (LLM-Powered) - Type 'exit' to quit\n")
while True:
    user_input = input("Describe your banking issue: ").strip()
    if user_input.lower() == "exit": break
    if not user_input: continue

    decision = analyze_banking_llm(user_input)
    
    if decision["create"]:
        res = create_banking_ticket(user_input, decision["category"], decision["priority"])
        print("\n--- 🏦 SECURE TICKET CREATED ---")
        print(f"Ticket ID : {res['id']}")
        print(f"Category  : {res['category']}")
        print(f"Priority  : {res['priority']}")
        print(f"Reference : {res['issue']}")
        print("--------------------------------\n")
    else:
        print("🤖 AI: No ticket required. Most login issues can be solved by resetting your password.\n")
```

### Create a  Weather Tool MCP Server that any AI agent can use with sample use case

```python
import requests
import json

class WeatherMCPServer:
    def __init__(self):
        self.weather_url = "https://api.open-meteo.com/v1/forecast"
        self.geo_url = "https://geocoding-api.open-meteo.com/v1/search"

    def get_tools(self):
        return [
            {
                "name": "get_coordinates",
                "description": "Get latitude and longitude for a city name.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "city": {"type": "string", "description": "Name of the city (e.g., London)"}
                    },
                    "required": ["city"]
                }
            },
            {
                "name": "get_weather_forecast",
                "description": "Get 3-day temperature forecast using coordinates.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "lat": {"type": "number"},
                        "lon": {"type": "number"}
                    },
                    "required": ["lat", "lon"]
                }
            }
        ]

    def call_tool(self, name, args):
        if name == "get_coordinates":
            res = requests.get(self.geo_url, params={"name": args["city"], "count": 1}).json()
            if "results" in res:
                return {"lat": res["results"][0]["latitude"], "lon": res["results"][0]["longitude"]}
            
        elif name == "get_weather_forecast":
            params = {
                "latitude": args["lat"], "longitude": args["lon"],
                "daily": "temperature_2m_max,temperature_2m_min", "timezone": "auto", "forecast_days": 3
            }
            data = requests.get(self.weather_url, params=params).json()
            days = data["daily"]
            forecast = []
            for i in range(3):
                forecast.append(f"{days['time'][i]}: High {days['temperature_2m_max'][i]}°C, Low {days['temperature_2m_min'][i]}°C")
            return "\n".join(forecast)
        
        return "Error: Tool or Location not found."

weather_server = WeatherMCPServer()
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv('GROQ_API_KEY'), base_url="https://api.groq.com/openai/v1")

def weather_mcp_agent(user_prompt):
    # Maintain a message history for multi-step tool use
    messages = [
        {"role": "system", "content": "You are a Weather Agent. Use 'get_coordinates' first if the user provides a city name, then use 'get_weather_forecast'."},
        {"role": "user", "content": user_prompt}
    ]

    # Loop to allow for multiple tool calls (Chain of Thought)
    for _ in range(3): 
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=messages,
            tools=[{"type": "function", "function": t} for t in weather_server.get_tools()],
            tool_choice="auto"
        )
        
        msg = response.choices[0].message
        messages.append(msg)

        if msg.tool_calls:
            for tool_call in msg.tool_calls:
                name = tool_call.function.name
                args = json.loads(tool_call.function.arguments)
                print(f"🛰️ Calling Server: {name}({args})")
                
                result = weather_server.call_tool(name, args)
                messages.append({
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": name,
                    "content": str(result)
                })
        else:
            return f"\n🤖 AI FINAL RESPONSE:\n{msg.content}"

# ============================================
# RUN LOOP
# ============================================
print("🌍 Smart Weather MCP Assistant (type 'exit')")
while True:
    query = input("\nEnter city for 3-day forecast: ")
    if query.lower() == 'exit': break
    print(weather_mcp_agent(query))
```
