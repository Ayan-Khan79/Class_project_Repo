```python
import os
import gradio as gr
from openai import OpenAI
from dotenv import load_dotenv

# 1. Configuration
load_dotenv()
endpoint = "https://ayan-resource.services.ai.azure.com/openai/v1"
deployment_name = "gpt-oss-120b"
Azure_Api_key = os.getenv('AZURE_API_KEY')

client = OpenAI(
    base_url=endpoint,
    api_key=Azure_Api_key
)

# 2. CGI Agent Instructions
SYSTEM_INSTRUCTIONS = """
You are the Clinical Guidance Intelligence (CGI) Agent.
Protocol: 1. Triage 2. Context Check 3. Evidence Retrieval.
Strict Guardrails: No Diagnosis, No Prescription.
Emergency Protocol: If an emergency is detected, respond ONLY with:
'IMMEDIATE ACTION REQUIRED: Call 911 or go to the ER immediately.'
Always include the Mandatory Disclosure Footer.
"""

# 3. Chat Logic
def healthcare_chat(user_input):
    try:
        completion = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {"role": "system", "content": SYSTEM_INSTRUCTIONS},
                {"role": "user", "content": user_input},
            ]
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"Connection Error: {str(e)}"

# 4. Gradio Notebook Deployment
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🏥 Analyst Training: Clinical AI Guardrails")
    with gr.Row():
        user_msg = gr.Textbox(label="Patient Query", placeholder="Input symptoms here...")
        output_text = gr.Textbox(label="Agent Response", interactive=False)
    
    submit_btn = gr.Button("Run Analysis", variant="primary")
    submit_btn.click(fn=healthcare_chat, inputs=user_msg, outputs=output_text)

# Launch with inline=True for Jupyter
demo.launch(inline=True, height=500,share=True)
```
